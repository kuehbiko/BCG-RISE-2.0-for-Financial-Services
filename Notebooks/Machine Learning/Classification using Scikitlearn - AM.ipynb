# <a id='toc2_'></a>[1. Logistic Regression <a id='p2' />](#toc0_)
<br>
 
Logistic regression is an extension of linear regression.<br/>
__When the dependent variable is dichotomous or binary in nature, we cannot use simple linear regression.__ <br>

__Logistic Regression__ is a mathematical model used to estimate the probability of an event occurring given some previous data. It's a __Classification Technique__.

-  It is still a regression technique but is commonly used for classification task.The probabilities are still continuous value outputs, it is the introduction of a threshold cut-off that renders it a classification task

- __Logistic regression outputs probabilities at its core functioning and doesn't output labels directly.__
- After probabilities are computed from the model, we can decide upon a threshold cut-off for classifying anything as success(1) or failure(0). <br> 

<b><u>Example</u></b>
- <mark>For instance, suppose we have customer loan data and we want to classify whether this particular customer will default on loan or not.</mark>
- <mark>Here, we have modeled the problem as a classification problem (yes/no), hence, Logistic Regression can be used here.</mark>

__In this training notebook, we will focus on classification using logistic regression. So, let's deep dive below!__


### <a id='toc2_1_1_'></a>[1.1 Why not Linear Regression? <a id='p2.1' />](#toc0_)

- Linear regression model can generate the predicted probability as any number ranging from negative to positive infinity, whereas probability of an outcome can only lie between 0< P(x)<1.<br/>
<mark>So it simply interpolates between the points, and you cannot interpret it as probabilities.</mark> <br> <br>


- Also, a linear regression model extrapolates and gives you values below zero and above one.<br/>
This is a good sign that there might be a smarter approach to classification. <br> <br>

- Since the predicted outcome is not a probability, but a linear interpolation between points, there is no meaningful threshold at which you can distinguish one class from the other.

![linvslog](images/linvslog.jpg)


### <a id='toc2_1_2_'></a>[1.2 Types of Logistic Regression <a id='p2.2' />](#toc0_)

<br>

<b><u> Binomial Logistic Regression</u></b>

As we saw earlier, this is the most common type of logistic regression technique in which binary outputs are considered (yes/no).
- Here, the dependent variable is a binary variable (discrete) and independent variables(s) can be either continuous or discrete
- <mark>An example could be to classify someone as a target customer for a company based on his/her profile and past data.</mark> 

__We will look into this in more detail in coming sections.__

<b><u> Multinomial Logistic Regression</u></b>

__Multinomial Logistic Regression__, or __Multiclass Classification__, is a technique where the target variable has more than 2 categories to be assigned.<br/>
In other words, our output is no longer a dichotomous variable. For eg, - Image classification.

Here, a common approach to use is __One vs All Approach__.<br/>
In this approach, each problem of label classification is treated as a single binary classification problem.<br/>
And the label with the highest predicted probability for positive events gets assigned to that instance.

<b><u> Example </u></b> <br> <br>
Suppose we have three labels for image classification - (dog, cat, mouse).
- <mark>Then, this approach will treat every problem as whether this is a dog or not, cat or not, etc.</mark>
- Therefore, we have three binary classification problems.
- And we know that logistic regression at its core outputs probabilities, hence, the label with the highest predicted probability (confidence in label) will be the winning candidate for that instance.


\* More reference in the External Link Library

    
<b><u> Ordinal Logistic Regression</u></b>

__Ordinal Logistic Regression__ is an ordinal regression model. i.e, a regression model for ordinal dependent variables.<br/>
In other words, if the categories to be assigned are in an ordered manner, the ordinal logistic regression may be used.<br/>
It can be thought of as an extension of the logistic regression model that applies to dichotomous dependent variables, allowing for more than two (ordered) response categories.

<b><u>Example</u></b> <br> <br>
<mark>If one question on a survey is to be answered by a choice among "poor", "fair", "good", and "excellent", and the purpose of the analysis is to see how well that response can be predicted by the responses to other questions, some of which may be quantitative, then ordered logistic regression may be used.</mark>

# <a id='toc3_'></a>[2. Introduction to Logistic Regression <a id='p3' />](#toc0_)

## <a id='toc3_1_'></a>[2.1 Assumptions <a id='p3.1' />](#toc0_)

Assumptions are basically the information that a model assumes about the data before it is entered for evaluation into it. A model is nothing but an equation that outputs a value when supplied with the particular predictor variable(s). <br>

So the model assumes that there is a certain distribution and relationship between variables followed by the data.

For instance, linear regression assumes a linear relationship between variables and, therefore, expects certain kind of properties and relationship constraints among the variables in the dataset as well. Hence, the end results will be useful & conclusive too when these assumptions are accounted for in the process pipeline.

Following are the assumptions behind Logistic regression:

1. Binary logistic regression requires the __dependent variable__ to be __binary__ and Ordinal logistic regression requires the dependent variable to be __ordinal__. 

2. It requires the __observations to be independent of each other__. In other words, the observations should not come from repeated measurements or matched data. 

3. It requires there to be __little or no multicollinearity among the independent variables__. This means that the independent variables should not be too highly correlated with each other.

4. It assumes __linearity of independent variables and log odds__. Although this analysis does not require the dependent and independent variables to be related linearly, it requires that the independent variables are linearly related to the log odds.

5. It typically __requires a large sample size__. A general guideline is that we need at minimum of 10 cases with the least frequent outcome for each independent variable in our model. For example, if we have 5 independent variables and the expected probability of our least frequent outcome is .10, then we would need a minimum sample size of 500 (10*5 / .10).

## <a id='toc3_2_'></a>[2.2 Binomial Logistic Regression <a id='p3.2' />](#toc0_)

The Logistic Regression model works on the fundamental _logit function_.<br/>
<mark>It models the log-odds for the value "1" (A success case) as a linear combination of one or more independent variables (features/predictors).</mark><br/>
Therefore, a simple logistic model has the form- <br>

![binomial_logit_formula](images/binomial_logit_formula.png)

where $\alpha$ is the __Y-Intercept__ or value of the function when all predictors are equal to zero, <br>
and $\beta$ is the __Regression Coefficient__ - the parameter estimated to model the relationship between outcome and predictor variable(s) <br>

Taking the antilog of Equation 1 on both sides, one derives at an equation to predict the probability of the occurrence of the outcome of interest as follows: <br>

![binomial_prob_formula](images/binomial_prob_formula.png)

where e = 2.71828 is the base of the system of natural logarithms.
<br><br>
X can be categorical or continuous, but Y is always categorical. According to Equation 1, __the relationship between logit (Y) and X is linear__.<br/>
Yet, according to Equation 2, __the relationship between the probability of Y and X is nonlinear__.<br/>
For this reason, the natural log transformation of the odds in Equation 1 is necessary to make the relationship between a categorical outcome variable and its predictor(s) linear.


<b><u> Null Hypothesis </u></b>

- Within the framework of inferential statistics, the null hypothesis states that β equals zero, or there is no linear relationship in the population.
- Rejecting such a null hypothesis implies that a linear relationship exists between X and the logit of Y.
- Hence, the resulting parameters and their predicted outcomes will determine whether the null hypothesis has been rejected or accepted based on the final evaluation metrics of model.

<b><u> Interpreting Coefficients </u></b>

- The value of the coefficient β determines the direction of the relationship between X and the logit of Y
- When β is greater than zero, larger (or smaller) X values are associated with larger (or smaller) logits of Y
- Conversely, if β is less than zero, larger (or smaller) X values are associated with smaller (or larger) logits of Y.

<b><u> Output </u></b>

- Logistic Regression at its core outputs probabilities where the default threshold in binary case is 0.5 for positive response.
- __If a predictor is binary, then the odds ratio is equal to e, the natural logarithm base, raised to the exponent of the slope β (e<sup>β</sup>)__.
- However, the threshold is subjective in nature depending upon the case at hand and can be changed, since probability is the main output and label assignment can be done easily on the user end.

## <a id='toc3_3_'></a>[2.3 Parameter Estimation <a id='p3.3' />](#toc0_)

We have seen the inner working of Logistic Regression and how it functions under the hood. Following is the equation again for your reference:

![binomial_logit_formula](images/binomial_logit_formula.png)

Here, log-odds is the target variable, $\alpha$ is the intercept, X are the feature variables and $\beta$ are the parameters.<br/>
We have just seen their relationship and interpretation, not how these parameters are being computed/estimated by the model.<br/>
There are several methods for deriving them and the process of estimating these parameters is called __Parameter Estimation__.

Parameters cannot be estimated accurately as well as easily with the manual approach. Since there are intricacies involved with large number of features as well as computational efficiencies on approaches, __optimized methods and solvers have been developed for this purpose to better estimate parameters.__ <br>

\* More reference in the External Link Library


<b><u> Techniques of Parameter Estimation </u></b>

- Different approaches are used for parameter estimation and __Model Fitting__ such as <b>Maximum Likelihood Estimation (MLE)</b> and <b>Iteratively Re-weighted Least Squares (IRLS).</b>

- <mark>However, α and βs are typically estimated by the Maximum Likelihood method</mark>, which is preferred over the Weighted Least Squares approach, since the Maximum Likelihood method is designed to maximize the likelihood of reproducing the data given the parameter estimates. <br>




### <a id='toc3_3_1_'></a>[2.3.1 Maximum Likelihood Estimation (MLE) <a id='p3.3.1' />](#toc0_)

MLE (Maximum Likelihood Estimation) is a method of estimating the parameters of a statistical model, given observations, by maximizing a likelihood function. The resulting arguments at the maximum are called __maximum likelihood estimates__, which is also abbreviated as MLE.<br/>
The parameter values are found such that they maximize the likelihood that the process described by the model produced the data that were actually observed. <br>

# <a id='toc4_'></a>[3. Binary Classification Metrics <a id='p8.1' />](#toc0_)

## <a id='toc4_1_'></a>[3.1 Confusion Matrix <a id='p8.1' />](#toc0_)

It is a matrix which allows the visualization of the performance of an algorithm, typically a supervised learning one, in which columns labels represent __Actual Values__, whereas the row labels represent __Predicted Values__. It shows the number of instances in each square computed by our model.

For instance, in reference to image below:
1. __TP (True Positive)__: Number of instances where the actual value and predicted value both are Yes
2. __FN (False Negative)__: Number of instances where the predicted value is No but actual value is Yes. It is also called __Type 2 Error__
3. __FP (False Positive)__: Number of instances where the predicted value is Yes but actual value is No. It is also called __Type 1 Error__
4. __TN (True Negative)__: Number of instances where the actual value and predicted value both are No. 

__confusion_matrix()__ function is used from scikit-learn for the same.

![confusion_matrix](images/confusion_matrix.png)

## <a id='toc4_2_'></a>[3.2 Accuracy <a id='p8.1' />](#toc0_)


*Accuracy* represents the number of correctly classified data instances over the total number of data instances. The formula for it is:

<img src='images/Accuracy.jpg' width="300" height="100" align="center"/>

Accuracy may not be a good measure if the dataset is not balanced (both negative and positive classes have different number of data instances).

## <a id='toc4_3_'></a>[3.3 Precision <a id='p8.1' />](#toc0_)

*Precision* is a measure of how many of the positive predictions made are correct (true positives). The formula for it is:

<img src='images/precision.jpg' width="300" height="100" align="center"/>

*Precision* should ideally be 1 (high) for a good classifier. Precision becomes 1 only when the numerator and denominator are equal i.e TP = TP +FP, this also means FP is zero. As FP increases the value of denominator becomes greater than the numerator and precision value decreases (which we don’t want).

## <a id='toc4_4_'></a>[3.4 Recall <a id='p8.1' />](#toc0_)

*Recall* is a measure of how many of the positive cases the classifier correctly predicted, over all the positive cases in the data. It is sometimes also referred to as Sensitivity. The formula for it is:

<img src='images/Recall.jpg' width="300" height="100" align="center"/>

*Recall* should ideally be 1 (high) for a good classifier. Recall becomes 1 only when the numerator and denominator are equal i.e TP = TP +FN, this also means FN is zero. As FN increases the value of denominator becomes greater than the numerator and recall value decreases (which we don’t want).

## <a id='toc4_5_'></a>[3.5 F1 Score <a id='p8.1' />](#toc0_)

*F1-Score* is a measure combining both precision and recall. It is generally described as the harmonic mean of the two. Harmonic mean is just another way to calculate an “average” of values, generally described as more suitable for ratios (such as precision and recall) than the traditional arithmetic mean. The formula used for F1-score in this case is:

### <a id='toc6_3_2_'></a>[5.3.2 Multicollinearity <a id='p5.3.3' />](#toc0_)

<br>
__Multicollinearity__ is a phenomenon where one independent variable can be explained with the help of one or more other independent variables present in the dataset.<br/>
In other words, it refers to a situation in which two or more explanatory variables in a multiple regression model are highly linearly related. <br> 

<b><u>Effects</u></b>

- This can induce a problem as bias/redundancy will be introduced in our model on account of wavering coefficients since it violates one of the basic assumptions of Logistic Regression.
- As a result, it becomes difficult to estimate the actual contribution of a feature in our model.

In practice, we rarely face perfect multicollinearity in a data set. It can lead to the following problems as well:
1. Estimates for regression coefficients can be unreliable
2. Tests of significance for regression coefficients can be misleading

<b><u>Quantifying Multicollinearity</u></b>

__The most common way to measure multicollinearity is through Variance Inflation Factor (VIF).__
- The __Variance Inflation Factor (VIF)__ quantifies the extent of correlation between one predictor and the other predictors in a model.
- Higher values signify that it is difficult to accurately assess the contribution of predictors to a model.
- It can be applied to any type of predictive model (e.g., CART, or deep learning).

<b><u>Interpreting VIF Values</u></b>

- <span style="background-color: #00ffff">The higher the value, the greater the correlation of the variable with other variables.</span>
- The higher the VIF, the more the standard error is inflated, and the larger the confidence interval and the smaller the chance that a coefficient is determined to be statistically significant.
- A value of 1 means that the predictor is not correlated with other variables.
- Values of more than 4 or 5 are sometimes regarded as being moderate to high, with values of 10 or more being regarded as very high.

But, these numbers are just rules of thumb.
- In some contexts a VIF of 2 could be a great problem (e.g., if estimating price elasticity)
- In straightforward predictive applications very high VIFs may be unproblematic.

So, the call is subjective in nature and depends on the situation at hand.

<mark>One of the features from pairs with high correlation should ideally be dropped to avoid the issue of multicollinearity.<br/>

We set a threshold for absolute correlation value above which we may drop one of the features.

```python
# Here, we set a threshold of abs(0.8)
threshold = 0.8
```

```python
from collections import defaultdict

adult_corr = adult.corr()

flag = False

corr_dict = defaultdict(list)

for row in adult_corr.index:
    for col in adult_corr.columns:
        if (col!=row) and (abs(adult_corr.loc[row, col]) >= threshold):
            flag = True
            corr_dict[row].append(col)

if flag:
    print('High correlation present!')
    print(corr_dict)
else:
    print('No high correlation present!')
```

<b>There are no pairs with correlation above our threshold of 0.8.</b>

Let's look at variance inflation factor:

```python
#VIF
#remove/encode all categorical variables prior to running this
from statsmodels.stats.outliers_influence import variance_inflation_factor

adult_wo_income = adult.drop(['income'],axis=1)

for i,k in enumerate(adult_wo_income.columns):
    print(i+1,'. ', k,': ', round(variance_inflation_factor(adult_wo_income.values, i),2),sep='')
```

There are various features with high VIF values.

<b>Let's choose a threshold of 5 and check which variables should be excluded from the analysis.<br/></b>
Below function drops the variable with highest VIF (above threshold) then re calculates the VIF values.<br/>
The process is repeated till all variables have a VIF below the threshold.

```python
#VIF function import
from statsmodels.stats.outliers_influence import variance_inflation_factor    

#define function to calculate VIF for all columns

def calculate_vif_(df, thresh=5):
    """
    The function calculates VIF for all columns, then drops the column with highest VIF above the threshold.
    The process is then repeated till there are no columns with VIF above the threshold.
    Function outputs the list of columns to be dropped.
    The columns are NOT actually dropped from the input dataframe
    Input Dataframe should only have numeric columns
    """
    variables = list(range(df.shape[1]))
    drop_cols = list()
    dropped = True
    
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(df.iloc[:, variables].values, ix) for ix in range(df.iloc[:, variables].shape[1])]

        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print(df.iloc[:, variables].columns[maxloc])
            drop_cols.append(df.iloc[:, variables].columns[maxloc])
            del variables[maxloc]
            dropped = True

    print('---')
    print('Columns to be dropped:')
    print(drop_cols)
    return drop_cols
    
drop_cols= calculate_vif_(adult_wo_income)
```

```python
#dropping above columns
adult.drop(drop_cols,axis=1,inplace=True)
```

#### <a id='toc8_2_1_1_'></a>[7.2.1.1 p-value <a id='p7.2.1.1' />](#toc0_)

<br>

Feature Significance is an important step in any model building process and __Statistical Criteria__ are in place to evaluate our model so that we can choose which features are the most significant contributors to our model.

All __hypothesis tests__ ultimately use a p-value to weigh the strength of the evidence.<br/>
The distance between the data and the model prediction is measured using a test statistic (such as a t-statistic or a chi-squared statistic).

__The p-value is the probability of obtaining results at least as extreme as the observed results, assuming that the null hypothesis is true.__ 

<b><u>Null Hypothesis</u></b>

This is the default hypothesis that we assume about the model and want to test for its correctness in terms of modeling.
- The null hypothesis for logistic regression is that no linear relationship exists between the features and the logit of y (target variable), i.e., $\beta$<sub>i</sub> = 0 for all values of i (features)

We'll test this Null Hypothesis through p-values later.

<b><u>Interpreting p-values</u></b>

The p-value is a number between 0 and 1 and interpreted in the following way:
- A small p-value (typically ≤ 0.05) indicates strong evidence against the null hypothesis, so you reject the null hypothesis 
- A large p-value (> 0.05) indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis 
- p-values very close to the cutoff (0.05) are considered to be marginal (could go either way). Always report the p-value so your readers can draw their own conclusions 
- _<span style="background-color: #00ffff"> Generally speaking, the threshold for p-value significance test is taken to be 0.05 most often. That means if the p-value of test statistic under consideration is greater than 5%, it states that we should not reject the null hypothesis, otherwise reject it. </span>_

<br>

<b><u>What is Cross-validation?</u></b>

Cross-validation is a technique used in statistics for model validation in order to see how well the model generalizes to an independent dataset.

The goal is to estimate the expected level of fit of a model to a data set that is independent of the data that were used to train the model.

It then points out how accurately will a model predict in unseen conditions, and flags problems such as Overfitting or Selection Bias, among others.

<b><u>Overfitting & Selection Bias</u></b>

- __Overfitting__:
 - It is the characteristic of a model that has been "too much accustomed" to the data on which it was trained, in sense that it doesn't generalize properly to unseen dataset.
 - It has learned way too many patterns that might just be _noise_ present in the dataset and assumed it to be its underlying model structure.

- __Selection Bias__:
 - It is the bias introduced by the selection of individuals, groups or data for analysis in such a way that proper randomization is not achieved
 - This means that the sample obtained is not representative of the population intended to be analyzed.

<b><u>Choice of Threshold</u></b>:

- Choosing threshold for assigning labels in a classification problem is an art in itself and it depends on __Business Context__.

<b><u>Business Sense</u></b>:
- We know that there will be some error associated with every model that we use for predicting the true class of the target variable. This will result in False Positives and False Negatives.
- There is no hard rule that says what should be minimized in all the situations. It purely depends on the business needs and the context of the problem you are trying to solve.
- Based on that, we might want to minimize either False Positives or False Negatives.

Let's understand this using an example:
- __Minimizing False Negatives__-A false negative, which means predicting that cancer is not present when it actually is, can result in a delay in treatment, which can lead to poorer outcomes, particularly for aggressive types of cancer where early detection and treatment are critical.

    
- __Minimizing False Positives__-A false positive, which means predicting that cancer is present when it is not, can lead to unnecessary stress, additional testing, potential complications from those tests, and unnecessary treatments.

Given these implications, many would argue that it's more crucial to minimize false negatives in cancer detection because missing an actual case of cancer (false negative) could lead to delayed treatment and potentially reduce survival rates. Early detection of cancer is often key to improving patient outcomes.

Therefore, it is a very subjective call and depends on business case at hand. It also depends on whether someone wants to follow a __conservative approach__ (very low threshold) or __aggressive approach__ (threshold cut-off on higher end).
