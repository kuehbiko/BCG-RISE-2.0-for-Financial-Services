# <a id='toc2_'></a>[1. Decision Tree Concepts & Introduction<br>](#toc0_)

Decision tree is a type of supervised learning algorithm that can be used in both regression and classification problems. It works for both __categorical__ and __continuous__ input and output variables.Decision tree looks like a flowchart like tree structure where an internal node represents feature(or attribute), the branch represents a decision rule, and each leaf node represents the outcome.

Important terminologies in decision trees

-  __Root Node__ represents the entire population or sample. It further gets divided into two or more homogeneous sets.

- Splitting is a process of dividing a node into two or more  __sub-nodes__.

- When a  __sub-node__ splits into further sub-nodes, it is called a Decision Node.

- Nodes that do not split is called a  __Terminal Node__ or a  __Leaf__.

- When you remove sub-nodes of a decision node, this process is called  __Pruning__. The opposite of pruning is  __Splitting__.

- A sub-section of an entire tree is called  __Branch__.

- A node, which is divided into sub-nodes is called a parent node of the sub-nodes; whereas the sub-nodes are called the  __child__ of the parent node.

## <a id='toc2_2_'></a>[1.2 Regression Tree <br>](#toc0_)

- One way to make predictions in a regression problem is to divide the predictor space (i.e. all the possible values for $X_1$, $X_2$,…, $X_p$) into distinct regions, say $R_1$, $R_2$,…,$R_k$ (terminal nodes or leaves).
- Then, for every $X$ that falls into a particular region (say $R_j$) we make the same prediction.
- Suppose we have two regions $R_1$ and $R_2$ with $Y_{1} = 10$ and $Y_{2} = 20$, respectively.
- For any value of $X$ such that $X \in {R}_1$, we would predict $10$. For any value of $X$ such that $X \in {R}_2$, we would predict $20$.
    
<img src="./Images/reg1.jpg" width="200" height="200" align="center"/>
    
- In this illustrative example above , we have two predictors and five distinct regions. Depending on which region the new $X$ comes from, we would make one of five possible predictions for $Y$.
- So find boxes $R_1$, . . . , $R_J$ that minimize the $\text{RSS}$
    
<img src="./Images/reg2.jpg" width="200" height="200" align="center"/>
    
where $\hat{y}_{R_j}$ is the mean response value of all training observations in the $R_j$ region
    
__STEPS__

1) Consider all predictor $X_p$ and all the all possible values of the cut points $s$ for each of the predictors. Choose the predictor and cut point such that it minimizes the $\text{RSS}$ <br>
2) Repeat 1 but only consider the sub-regions <br>
3) Stop: node contains only one class or  node contains less than n data points or max depth is reached

<font><font style="font-family:Trebuchet MS;">

## <a id='toc2_3_'></a>[1.3 Classification Tree <br>](#toc0_)

- Very similar to regression except that it is used to predict a qualitative response rather than a quantitative one 
- In regression trees we use the mean response of the training observations, for classification trees we use most commonly occurring class
- $\text{RSS}$ cannot be used as a criterion for making the binary splits.

__Classification error rate__

## <a id='toc2_4_'></a>[1.4 Summary Regression Tree vs Classification Tree<br>](#toc0_)
    
Both the trees work almost similar to each other, let’s look at the primary differences & similarities between classification and regression trees:

- Regression trees are used when dependent variable is continuous. Classification trees are used when dependent variable is categorical.
- In case of regression tree, the value obtained by terminal nodes in the training data is the mean response of observation falling in that region. Thus, if an unseen data observation falls in that region, we’ll make its prediction with mean value.
- In case of classification tree, the value (class) obtained by terminal node in the training data is the mode of observations falling in that region. Thus, if an unseen data observation falls in that region, we’ll make its prediction with mode value.
- Both the trees divide the predictor space (independent variables) into distinct and non-overlapping regions. For the sake of simplicity, we can think of these regions as high dimensional boxes or boxes.
- Both the trees follow a top-down greedy approach known as recursive binary splitting. We call it as top-down because it begins from the top of tree when all the observations are available in a single region and successively splits the predictor space into two new branches down the tree. It is known as greedy approach because, the algorithm cares (looks for best variable available) only the current split, and not about future splits which will lead to a better tree.
- This splitting process is continued until a user defined stopping criteria is reached. For example: we can tell the the algorithm to stop once the number of observations per node becomes less than 50.
- In both the cases, the splitting process results in fully grown trees until the stopping criteria is reached. But, the fully grown tree is likely to over-fit data, leading to poor accuracy on unseen data. This bring pruning. Pruning is one of the technique used to tackle over-fitting.

### <a id='toc2_5_3_'></a>[1.5.3 Splitting Criterion in Classification Trees<br>](#toc0_)

- __Classification Error Rate:__ Here "hit rate" is used which is the fraction of training observations in a particular region that don't belong to the most widely occuring class. The error is given by this equation:

   $E = 1 - argmax_{c}(π^mc)$

   in which π^mc represents the fraction of training data in region Rm that belong to class c.
   

- __Gini Index:__ The Gini Index is an alternative error metric that is designed to show how "pure" a region is. "Purity" in this case means how much of the training data in a particular region belongs to a single class. If a region contains data that is mostly from a single class then the Gini Index value will be small. In simple words, small value means that Gini impurity is small (i.e., purity is high)

<img src="./Images/gini.jpg" width="220" height="120" align="center"/><br>

- __Cross-Entropy:__ A third alternative, which is similar to the Gini Index, is known as the Cross-Entropy or Deviance:The cross-entropy will take on a value near zero if the π^mc’s are all near 0 or near 1. Therefore, like the Gini index, the cross-entropy will take on a small value if the mth node is pure. In fact, it turns out that the Gini index and the cross-entropy are quite similar numerically.

<img src="./Images/entropy.jpg" width="220" height="120" align="center"/><br>

where $p_{i}$ is defined as the proportion of subsamples that belong to a certain target class.

When building a classification tree, either the Gini index or the cross-entropy are typically used to evaluate the quality of a particular split, since they are more sensitive to node purity than is the classification error rate. Any of these 3 approaches might be used when pruning the tree, but the classification error rate is preferable if prediction accuracy of the final pruned tree is the goal.

__Accuracy__

Accuracy = (TP + TN) / Total examples

__Precision__

Also called the positive predictive value, this metric is the number of correct predictions divided by the number of correct predictions plus false positives.Precision helps to determine when the model predicts positive how often it is correct.

Precision=TP/(TP+FP)

F1 is a helpful measure of a test’s accuracy. It is a consideration of both precision and recall, and an F1 score is considered perfect when at 1 and is a total failure when at 0, as defined by sklearns

### <a id='toc2_5_8_'></a>[1.5.8 ROC Curve<br>](#toc0_)
    
The ROC curve is a valuable tool to assess the performance of a classification model across different discrimination thresholds and to compare different models. 
    
It shows the trade-off between the true positive rate (sensitivity) and the false positive rate (1 - specificity) as the discrimination threshold of the model is varied. 
    
Here's how to interpret and create an ROC curve:

* __True Positive Rate (TPR):__ This is also known as sensitivity or recall. It is the ratio of correctly predicted positive instances to the total actual positive instances.
<br>    
* __False Positive Rate (FPR):__ This is the ratio of incorrectly predicted negative instances to the total actual negative instances. It is also equal to 1−Specificity.     

* __Threshold__ <br>
A threshold is simply a cutoff for the model’s output. The output for a binary classifier is a probability, and if the output is higher than the threshold(for example, 0.50), it will be assigned to 1, else 0. <br>   
    
If the threshold was to be set at 0.00, then the classifier will predict every observation as 1. This will therefore result FPR and TPR to be 1 because there will be no negatives, regardless of true or not(FP / (FP + 0) = 1 and TP / (TP + 0) = 1).

If threshold was to be set at 1.00, then every observation will be predicted as 0 (0 / (0 + TN) = 0 and 0 / (0 + FN) = 0) and therefore the TPR and FPR will be 0.    
    
__How to create a ROC curve?__
<br>
* Train a classification model 
* Obtain the predicted probabilities or scores from the model for each instance of the dataset
* Sort the instances based on these scores in descending order
* Vary the discrimination threshold from 0 to 1. For each threshold, calculate the TPR and FPR
* Plot the TPR against the FPR to create the ROC curve

Let's look at the graphical representation of ROC curve  
    
<img src="./Images/roc-curve.png" width="400" align="center"/><br>


__Interpreting the ROC Graph:__

* Points on the ROC curve represent different threshold settings, showing the TPR and FPR at each point.
* The diagonal dashed line (FPR = TPR) represents the ROC curve for a random classifier, where the model's performance is no better than chance.
* The further the ROC curve is from the diagonal line toward the upper-left corner, the better the model's performance.
* If the curve is located in the top left corner of the plot, that indicates an ideal model i.e., a false positive rate of 0 and true positive rate of 1
* The choice of a threshold depends on the specific application's needs. A threshold choice depends on the desired balance between sensitivity and specificity.

### <a id='toc2_5_9_'></a>[1.5.9 AUC Score<br>](#toc0_)
It is a probability curve and AUC (Area Under Curve) represents degree or measure of separability. It tells us how much the model is capable of distinguishing between classes. Higher the AUC, better the model is at predicting 0s as 0s and 1s as 1s. 
A poor model has AUC near to the 0 which means it has worst measure of separability. It is predicting 0s as 1s and 1s as 0s. In simple words, AUC near 0 happens when the predictions are opposite of the ground truth. And when AUC is 0.5, it means model has no class separation capacity whatsoever. Means, the model is simply making random guesses when AUC is 0.5. <br>

A model with a higher ROC AUC (Area Under the Curve) generally indicates better overall performance.    

\* More reference in the External Link Library

```python
import sys
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds =roc_curve(y_test, y_pred_dtree)
roc_auc = auc(fpr, tpr)

def plot_roc_curve(fpr, tpr, auc, mod, xlim=None, ylim=None):
    """
    Purpose
    ----------
    Function creates ROC Curve for respective model given selected parameters.
    Optional x and y limits to zoom into graph

    Parameters
    ----------
    fpr:  Array returned from sklearn.metrics.roc_curve for increasing
         false positive rates
    tpr:  Array returned from sklearn.metrics.roc_curve for increasing
         true positive rates
    auc:  Float returned from sklearn.metrics.auc (Area under Curve)
    mod:  String represenation of appropriate model, can only contain the
         following: ['knn', 'rf', 'nn']
    xlim: Set upper and lower x-limits
    ylim: Set upper and lower y-limits

    Returns:
    ----------
    Returns plot of Receiving Operating Curve for specific model. Allowing user to
    specify x and y-limits.
    """
    mod_list = ['knn', 'rf', 'nn']
    method = [('Kth Nearest Neighbor', 'deeppink'),
          ('Random Forest', 'red'),
          ('Neural Network', 'purple')]

    plot_title = ''
    color_value = ''
    for i in range(0, 3):
        if mod_list[i] == mod:
            plot_title = method[i][0]
            color_value = method[i][1]

    fig, ax = plt.subplots(figsize=(10, 10))
    #ax.set_axis_bgcolor('#fafafa')

    plt.plot(fpr, tpr,
         color=color_value,
         linewidth=1)
    plt.title('ROC Curve For decision tree (AUC = {1: 0.3f}) \
          \nLoan Prediction Diagnostic'\
          .format(plot_title, auc))

    plt.plot([0, 1], [0, 1], 'k--', lw=2) # Add Diagonal line
    plt.plot([0, 0], [1, 0], 'k--', lw=2, color = 'black')
    plt.plot([1, 0], [1, 1], 'k--', lw=2, color = 'black')
    if xlim is not None:
        plt.xlim(*xlim)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.show()
    plt.close()
    
plot_roc_curve(fpr, tpr, roc_auc, 'rf',
           xlim=(-0.01, 1.05),
           ylim=(0.001, 1.05))
```

### <a id='toc2_5_10_'></a>[1.5.10 Hyper parameter Optimization<br>](#toc0_)
Now, let's create the model, starting with hyperparameter tuning. Here are the hyperparameters we will be tuning in this tutorial: 

-  __max_depth__: The maximum splits for all trees in the forest. The more the splits more flexible the model is.
-  __bootstrap__: An indicator of whether or not we want to use bootstrap samples when building trees.
-  __max_features__: The maximum number of features that will be used in node splitting — the main difference I previously mentioned between bagging trees and random forest. Typically, you want a value that is less than p, where p is all features in your data set.
-  __criterion__: This is the metric used to asses the stopping criteria for the decision trees.

The parameters helps to control regularization and model complexity
Utilizing the GridSearchCV functionality, let's create a dictionary with parameters we are looking to optimize to create the best model for our data. Setting the n_jobs to 3 tells the grid search to run three jobs in parallel, reducing the time the function will take to compute the best parameters. We have included the timer to see how long different jobs took; that led me to ultimately decide to use three parallel jobs.<br>

\* More reference in the External Link Library


```python
clf = DecisionTreeClassifier(random_state=42)
from sklearn.model_selection import train_test_split, GridSearchCV
np.random.seed(42)
start = time.time()

param_dist = {'max_depth': [2, 3, 4],
              'max_features': ['auto', 'sqrt', 'log2', None],
              'criterion': ['gini', 'entropy']}

cv_clf =GridSearchCV(clf, cv = 10,
                     param_grid=param_dist,
                     n_jobs = 3,scoring='accuracy')

cv_clf.fit(x_train,y_train)
print('Best Parameters using grid search: \n',
     cv_clf.best_params_)
end = time.time()
print('Time taken in grid search: {0: .2f}'.format(end - start))
```

<font><font style="font-family:Trebuchet MS;">
Once we find the best parameter combination, we can set the parameters to our model.

```python
# Set best parameters given by grid search
clf.set_params(criterion = 'gini',max_features = 'auto',max_depth = 2)
```

```python
clf.fit(x_train,y_train)
y_pred_cv=clf.predict(x_test)
```

## <a id='toc2_7_'></a>[1.7 What is  CHAID ?<br>](#toc0_)

CHAID, or Chi-square Automatic Interaction Detection, is a Classification Tree technique
that not only evaluates complex interactions among predictors, but also displays the modeling
results in an easy-to-interpret tree diagram. 

- The dependent variable must be categorical
- CHAID uses Chi-Squared test using contingency tables as a splitting algorithm
- CHAID splits into multiple groups

When the dependent variable we are trying to predict has only two values (e.g., mail responder
vs. non-responder), we generate a nominal CHAID model. In such a model, we are able to see
what proportion of each market segment consists of cases in the desired category of the
dependent variable (e.g., mail responders).

When the dependent variable is at least ordinal (i.e., the values can be arranged in some
meaningful order), then we generate an ordinal CHAID model. Customer dollar value is an
example of an ordinal dependent variable. In an ordinal model, each segment is assigned an
average value on the dependent variable (e.g., average dollar value).

### <a id='toc2_8_3_'></a>[1.8.3 Splitting Criterion in Regression Trees<br>](#toc0_)
    
- In decision tree regression, the splitting criterion is used to decide how to split the data at each internal node of the tree. 
The goal of the splitting criterion is to find the best feature and threshold to minimize the impurity 
or variance within each leaf node, resulting in more homogeneous groups of data points or simply more accurate predictions.
    
- In scikit-learn's DecisionTreeRegressor, the criterion parameter allows you to choose the function that measures the quality 
of a split when building the decision tree. 
Different criteria are used to evaluate the split at each node, and the best split is chosen based on the criterion 
that optimizes the tree's performance. Here's an explanation of the available criteria to choose further for the problem in-hand:

The splitting criteria for decision regression trees are as follows:

- __Mean Squared Error (MSE) or "squared_error":__ MSE measures how close our predictions are to the actual values. We take the difference between each predicted value and the actual value, square that difference, and then find the average of all these squared differences.<br>
    
    Mathematically, MSE = Σ(predicted - actual)^2 / number_of_data_points. <br>
    
    A lower MSE means our predictions are more accurate, as the squared differences are small.<br> 
Let's understand these criterias in a simpler way for the housing price prediction problem: <br>
Imagine you want to predict house prices. The MSE measures how well your predictions match the actual house prices.
It calculates the average of the squared differences between your predicted prices and the actual prices.
If the MSE is low, it means your predictions are close to the real prices, which is good because you're making accurate predictions. <br>
<br>    
    
- __Friedman's Mean Squared Error or "friedman_mse":__ Friedman's MSE is similar to regular MSE, but it considers the number of data points in the calculation. It scales the squared differences by the number of data points in the dataset. <br>
    
    Mathematically, Friedman's MSE = Σ(predicted - actual)^2 / (number_of_data_points * scaling_factor)  <br>
    
    It helps when we have a small dataset and want to make sure the model is not biased by the number of data points  <br>
Let's understand these criterias for the housing price prediction problem: <br>
This is similar to MSE, but it takes into account the number of houses in your dataset.
If you have a smaller dataset, this criterion might be better because it considers the size of your data when evaluating predictions.<br>
<br>
- __Mean Absolute Error (MAE) or "absolute_error":__ MAE also measures how close our predictions are to the actual values, but it uses the absolute value of the differences instead of squaring them. It finds the average of all these absolute differences.  <br>
<br>
    Mathematically, MAE = Σ|predicted - actual| / number_of_data_points  <br>
    
    MAE is useful when we want a more balanced view of how well our predictions are doing, and we don't want large errors to have too much impact.  <br>
Let's understand these criterias for the housing price prediction problem: <br>
Instead of squaring the differences like MSE, MAE takes the absolute value of the differences between your predictions and the actual prices.
MAE is useful when you want your model to be less sensitive to large prediction errors. It provides a more balanced view of how well your model is doing.<br>
<br>   
- __Poisson or "poisson":__ The Poisson criterion is suitable when we are predicting numbers that represent counts, like the number of houses sold in a month. It helps our model make predictions that are more suitable for counting data.
The specific mathematical formula depends on the details of the Poisson distribution, which may not be as intuitive for a layman audience.<br>
Let's understand these criterias for the housing price prediction problem: <br>
Imagine you are predicting the number of houses sold in a month. Poisson criterion is suitable when dealing with counting numbers, like the number of houses sold.
It helps your model make predictions more accurately based on the counting nature of the target variable.<br>    
<br>
For predicting housing prices, let's start with using the default Mean Squared Error (MSE) criterion. It helps ensure that our predictions are accurate, and we can easily understand how far off our predictions are from the actual prices. However, it's always good to experiment with different criteria and see which one performs best for your specific dataset and problem. Remember, the goal is to find the criterion that gives you the most accurate predictions for your house price predictions.

#### <a id='toc2_8_7_1_'></a>[1.8.7.1 R-squared<br>](#toc0_)

R-squared also known as Coefficient of determination is a statistical measure of how close the data are to the fitted regression line. It is also known as the coefficient of determination, or the coefficient of multiple determination for multiple regression.

The definition of R-squared is fairly straight-forward; it is the percentage of the response variable variation that is explained by a linear model. Or:

R-squared = Explained variation / Total variation

R-squared is always between 0 and 100%:

0% indicates that the model explains none of the variability of the response data around its mean. 100% indicates that the model explains all the variability of the response data around its mean. In general, the higher the R-squared, the better the model fits your data.

__Key Limitations of R-squared__

R-squared cannot determine whether the coefficient estimates and predictions are biased, which is why you must assess the residual plots.

R-squared does not indicate whether a regression model is adequate. You can have a low R-squared value for a good model, or a high R-squared value for a model that does not fit the data.

__Are Low R-squared Values Inherently Bad?__

No! There are two major reasons why it can be just fine to have low R-squared values.

In some fields, it is entirely expected that your R-squared values will be low. For example, any field that attempts to predict human behavior, such as psychology, typically has R-squared values lower than 50%. Humans are simply harder to predict than, say, physical processes.

Furthermore, if your R-squared value is low but you have statistically significant predictors, you can still draw important conclusions about how changes in the predictor values are associated with changes in the response value. Regardless of the R-squared, the significant coefficients still represent the mean change in the response for one unit of change in the predictor while holding other predictors in the model constant. Obviously, this type of information can be extremely valuable.

It measures the proportion of the variance in the target variable that the model explains or accounts for in its predictions

__Here's how the explained_variance_score is calculated step-by-step:__

- __Calculate the Residuals:__ Find the differences between the true target values (y_true) and the predicted target values (y_pred). These differences are called residuals.

- __Calculate the Variance of Residuals:__ Calculate the variance of the residuals. The variance represents the average of the squared differences from the mean. It quantifies how much the residuals spread out around the mean value.

- __Calculate the Variance of True Target Values:__ Calculate the variance of the true target values (y_true). This measures the variability in the actual target values.

- __Calculate the Explained Variance:__ The explained_variance_score is then calculated as 1 minus the ratio of the variance of residuals to the variance of true target values. The score ranges from negative infinity to 1. A score of 1 indicates that the model perfectly predicts the target variable, while lower scores indicate less accurate predictions.

    The explained_variance_score is a useful metric to assess the quality of regression models. A higher value indicates a better model, as it suggests that the model explains a larger proportion of the variance in the target variable. However, it's important to interpret this score in the context of the specific problem and the range of target variable values.

For a regression tree, feature importance is typically determined based on how much each feature contributes to reducing the impurity or variance within the tree

There are different methods to calculate feature importance for a regression tree:

__Gini Importance:__ For each feature, the Gini importance measures the total reduction of impurity (Gini index) that is achieved by splitting on that feature across all the nodes of the tree.

__Mean Decrease in Impurity (MDI):__ Similar to Gini importance, MDI calculates the average reduction of impurity at each split point for a given feature, and then averages this reduction across all trees in an ensemble (e.g., Random Forest).

__Mean Decrease in Accuracy (MDA):__ This method shuffles the values of a single feature in the test data and measures how much the accuracy (R-squared for regression) decreases. A larger drop in accuracy indicates a more important feature.

__Permutation Importance:__ Permutation importance is similar to MDA, but instead of shuffling the test data, it shuffles the values of a single feature in the training data and measures the drop in the evaluation metric (e.g., R-squared). This approach is computationally more efficient than MDA.


## <a id='toc2_9_'></a>[1.9 Advantages and Disadvantages of Decision Trees<br>](#toc0_)
The major advantage of using decision trees is that they are intuitively very easy to explain. They closely mirror human decision-making compared to other regression and classification approaches. They can be displayed graphically, and they can easily handle qualitative predictors without the need to create dummy variables.

However, decision trees generally do not have the same level of predictive accuracy as other approaches, since they aren't quite robust. A small change in the data can cause a large change in the final estimated tree i.e., Decision trees are prone to overfitting

By aggregating many decision trees, using methods like bagging, random forests, and boosting, the predictive performance of decision trees can be substantially improved.<br>
