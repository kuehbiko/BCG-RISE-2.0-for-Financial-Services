# 3. While Loop <a id='p3' />

A loop is used to repeat a block of commands multiple times. There are two ways to write a loop, one is a `for` loop and the other is a `while` loop. Typically, you use a `while` loop when looping is based on a conditional that will be modified during the loop.

A `while` loop is pretty simple, it's structure looks like:

    while expression:
        # statement(s)
        ...
        
and it continues until `a_condition` is false.


    
    


Figure Source: Geeksforgeeks.org

## 3.1 Break

To break out of a while loop in Python, you can use the `break` statement. When the break statement is encountered within a loop, it immediately terminates the loop and execution continues with the next statement after the loop.

It's important to note that using break will exit the innermost loop it is placed in. If you have nested loops and want to break out of an outer loop, you can use additional control structures like flags or labels.

Remember to use the break statement with caution and ensure that your loop termination conditions are properly defined to avoid creating infinite loops.

## 3.2 Continue

To skip the current iteration of a while loop in Python and continue with the next iteration, you can use the `continue` statement. When the continue statement is encountered within a loop, it immediately skips the remaining code in that iteration and moves to the next iteration.

It's important to note that using continue will skip the remaining code within the current iteration of the loop and move to the next iteration. Make sure that your loop conditions and logic are properly defined to avoid creating infinite loops.

# 4. For Loop <a id='p4' />

A `for` loop lets us repeat a set of commands defined number of times. The syntax for a `for` loop is just:

    for item in sequence:
        # do something with item
        ...

But what is a sequence?

There are lots of functions in Python that will actually return a sequence - they are called *iterators*. An iterator essentially provides the next element in the sequence each time we access it. 

An iterator could be a list, tuple, set, dictionary, range function, string etc.


Figure Source: Geeksforgeeks.org

Here, we will first use the `range` function as an iterator.

The `range()` function in Python is used to generate a sequence of numbers. It creates an immutable sequence of numbers starting from a specified start value (inclusive) and ending at a specified stop value (exclusive), with a specified step size.

The syntax of the range() function is as follows:

range(start, stop, step)

where:

start (optional): The starting value of the sequence. If not specified, the default value is 0.

stop (required): The end value of the sequence. The sequence will stop before reaching this value.

step (optional): The step size or increment between numbers in the sequence. If not specified, the default value is 1.
